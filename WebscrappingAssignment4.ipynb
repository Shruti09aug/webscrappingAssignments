```python
Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details:
A) Rank
B) State
C) GSDP(18-19)
D) GSDP(17-18)
E) Share(2017)
F) GDP($ billion)
Note: - From statisticstimes home page you have to reach to economy page through code.
```

```python
# Connecting to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
# Getting the webpage of mentioned url 

driver.get('http://statisticstimes.com/')
driver.implicitly_wait(3)
#ActionChains Required Library
from selenium.webdriver.common.action_chains import ActionChains
```

From the home page we have to select Economy and Inida option. 
we can see that Economy option is not clickable, we have to do mouseover action .  
We can perform mouseover action in Selenium webdriver in Python by using the ActionChains class. 
We have to create an object of this class and then apply suitable methods on it.
In order to move the mouse to an element, we shall use the move_to_element method and pass the element locator as a parameter. 
Then apply the perform method to actually perform this action. After hovering on the element, we can apply click action on it with the help of the click method.
ActionChains Required Library

```python
#object of ActionChains
a = ActionChains(driver)
#identify element
m = driver.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/button")
#hover over element
a.move_to_element(m).perform()
#identify sub menu element
n = driver.find_element(By.LINK_TEXT,"India")
# hover over element and click
a.move_to_element(n).click()
```

```python
#We need state wise GDP. So we will select "GDP of Indian States. With click() option."
driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[2]/ul/li[1]/a").click()
```

```python
#Creating empty list
Rank = []
State = []
GSDP_18_19 = []
GSDP_19_20 = []
Share = []
GDP_billions = []
```

```python
ra=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[1]")
for i in ra:
    Rank.append(i.text)
sa=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[2]")

for i in sa:
    State.append(i.text)
gsd=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[4]")

for i in gsd:
    GSDP_18_19.append(i.text)
gs=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[3]")

for i in gs:
    GSDP_19_20.append(i.text)
sha=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[5]")

for i in sha:
    Share.append(i.text)
bil=driver.find_elements(By.XPATH,"//div[@class='fwidth'][3]/div/div/table/tbody/tr/td[6]")

for i in bil:
    GDP_billions.append(i.text)
```

```python
#Creating the data frame for the scrapped data. 
Economy=pd.DataFrame({})
Economy['Rank']=Rank
Economy['State']=State
Economy['GSDP_18_19']=GSDP_18_19
Economy['GSDP_19_20']=GSDP_19_20
Economy['Share']=Share
Economy['GDP_billions']=GDP_billions
Economy
```

```python
Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
ASSIGNMENT
Note: - From the home page you have to click on the trending option from Explore menu through code.
```

```python
# Connecting to the webdriver
driver=webdriver.Chrome("chromedriver.exe")
# Getting the webpage of mentioned url 

driver.get('https://github.com/')
driver.implicitly_wait(3)
#ActionChains Required Library
from selenium.webdriver.common.action_chains import ActionChains
```

```python
Trending=driver.find_element(By.XPATH,"//a[@href='/trending']") #Looking for Trending
driver.get(Trending.get_attribute('href')) #Open Trending page


Repository_title = []
Repository_description = []
Contributors_count = []
Language_used = []
for i in range(1,26): #25 repository are trending so running the loop for 1 to 26
    link1 = driver.find_element(By.XPATH,"//article[@class='Box-row']["+str(i)+"]/h1/a") #Repository Title
    Repository_title.append(link1.text)
    try:
        link2 = driver.find_element(By.XPATH,"//article[@class='Box-row']["+str(i)+"]/p") #Repository Description
        Repository_description.append(link2.text)
    except NoSuchElementException as e:
        Repository_description.append("-") #If Description is not present add "-"
    
urls = []
contr=driver.find_elements(By.XPATH,"//article[@class='Box-row']/h1/a") #Collecting the URLS of each repository
for i in contr:
    urls.append(i.get_attribute('href')) #Added url in urls field
    
for i in urls:
    driver.get(i) #Open the URL one by one
  
    counter = 0
    for i in range(1,6):
        try:
            test = driver.find_element(By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/h2/a") #Searching for Contributors
            value = test.text
            if(value.find("Contributors") != -1): #If found than
                con = value.split(" ")            #Split value with space
                Contributors_count.append(con[1]) #Second value will be number of Contributors
            else:
                counter += 1
        except NoSuchElementException as e:
            counter += 1
            pass
    if(counter == 5):
        Contributors_count.append("-")           #If no Contributors for that Repository than add "-"
        
    count = 0
    for i in range(1,7):
        try:
            test = driver.find_element(By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/h2") #Searching for Language Used
            value = test.text
            string = ""
            if(value.find("Languages") != -1): #If Language found than
                lang = driver.find_elements(By.XPATH,"//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/ul/li/a/span[1]")
                for i in lang:
                    string = string + i.text + " " #Added all languages one by one
                Language_used.append(string)
            else:
                count += 1
        except NoSuchElementException as e:
            count += 1
            pass
    if(count == 6):
        Language_used.append("-")                 #If no Language found then add "-"
        
Github = pd.DataFrame({}) #DataFrame
Github['Repository Title'] = Repository_title
Github['Repository Description'] = Repository_description
Github['Contributors Count'] = Contributors_count
Github['Language Used'] = Language_used
Github
```
